```python
import game
import numpy as np
import matplotlib.pyplot as plt
```

```python
game = game.Game("dqn_agent", "cpu")


hyperparameter = {
  "lr_start": 1e-4,
  "lr_end": 1e-4,
  "batch_size": 128,
  "gamma": 0.9,
  "eps_start": 0.9,
  "eps_end": 1e-2
}


# loss_hist, performance_hist = game.train_agent(False, 100, 100, hyperparameter)  
loss_logs, test_score_logs = game.train_agent(False, 100, 100, hyperparameter) 
score = game.main(True)
print("Score: ", score)
```

```python
# Plot the performance and the loss of each episode
fig, axes = plt.subplots(nrows=1, ncols=1)

axes.plot(test_score_logs)
axes.set_title('Performance')
axes.set_xlabel('Epoch')
axes.set_ylabel('test_score_logs (Performance)')

plt.show()
```

```python
# Plot the performance and the loss of each episode
fig, axes = plt.subplots(nrows=1, ncols=1)

axes.plot(loss_logs)
axes.set_title('Loss')
axes.set_xlabel('Epoch')
axes.set_ylabel('Loss rate')
```
